# **2. 선형회귀 구현하기 예제 - 경사 하강법의 적용**

# 2.1 예측값으로 올바른 모델찾기

1. w와 b 초기화 하기

```python
w = 1.0
b = 1.0
```

2. 훈련데이터의 첫번째 샘플 얻기

```python
y_hat = x[0] * w + b
print(y_hat)
```

3. 타깃과 예측 데이터 비교하기

```python
print(y[0])
```

4. w 값 조절해 예측값 바꾸기

```python
w_inc = w + 0.1
y_hat_inc = x[0] * w_inc  + b
print(y_hat_inc)
```

5. w 변화율 구하기 - w 값 조정한 후 예측값 증가 정도 확인하기

```python
w_rate = (y_hat_inc - y_hat) / (w_inc - w)
print(w_rate)
print(x[0])
```

6. w (가중치) 업데이트 하기

```python
w_new = w + w_rate
print(w_new)
```

7. b(절편) 업데이트 하기

```python
b_inc = b + 0.1
y_hat_inc = x[0] * w + b_inc
print(y_hat_inc)
```

```python
b_rate = (y_hat_inc - y_hat) / (b_inc - b)
print(b_rate)
```

```python
b_new = b + 1
print(b_new)
```

#**2.2 오차역전파로 가중치와 절편을 더 적절하게 업데이트 하기**


1. 오차와 변화율을 곱하여 가중치 업데이트하기

```python
err = y[0] - y_hat
w_new = w + w_rate * err
b_new = b + 1 * err
print(w_new, b_new)
print(err)
```

2. 2번 샘플을 이용하여 w와 b 업데이트 하기

```python
y_hat = x[1] * w_new + b_new
err = y[1] - y_hat
w_rate = x[1]
w_new = w_new + w_rate * err
b_new = b_new + 1 * err
print(w_new, b_new)
```

3. 전체 샘플을 반복해서, w와 b 업데이트 하기

```python
for x_i, y_i in zip(x, y):
    y_hat = x_i * w + b
    err = y_i - y_hat
    w_rate = x_i
    w = w + w_rate * err
    b = b + 1 * err
print(w, b)
```

```python
print(err)
```

4. 데이터 분포도와 찾아낸 w와 b를 이용한 그래프 함께 그리기

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

5. 에러 에포크를 100번 반복하기

```python
for i in range(1, 100):
    for x_i, y_i in zip(x, y):
        y_hat = x_i * w + b
        err = y_i - y_hat
        w_rate = x_i
        w = w + w_rate * err
        b = b + 1 * err
```

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('./output/regression.png')
plt.show()
```

# **3. 뉴런 만들기 예제**

```python
class Neuron:
    
    def __init__(self):
        self.w = 1.0     # 가중치를 초기화합니다
        self.b = 1.0     # 절편을 초기화합니다
    
    def forpass(self, x):
        y_hat = x * self.w + self.b       # 직선 방정식을 계산합니다
        return y_hat
    
    def backprop(self, x, err):
        w_grad = x * err    # 가중치에 대한 그래디언트를 계산합니다
        b_grad = 1 * err    # 절편에 대한 그래디언트를 계산합니다
        return w_grad, b_grad

    def fit(self, x, y, epochs=100):
        for i in range(epochs):           # 에포크만큼 반복합니다
            for x_i, y_i in zip(x, y):    # 모든 샘플에 대해 반복합니다
                y_hat = self.forpass(x_i) # 정방향 계산
                err = -(y_i - y_hat)      # 오차 계산
                w_grad, b_grad = self.backprop(x_i, err)  # 역방향 계산
                self.w -= w_grad          # 가중치 업데이트
                self.b -= b_grad          # 절편 업데이트
```

```python
neuron = Neuron()
neuron.fit(x, y)
```

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * neuron.w + neuron.b)
pt2 = (0.15, 0.15 * neuron.w + neuron.b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('./output/neural network.png')
plt.show()
```
